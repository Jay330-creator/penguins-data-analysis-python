# Penguin Data Analysis (Python EDA)

This project analyzes penguin data to explore differences between species using data visualization and exploratory data analysis techniques.

## Objective
The goal of this project is to analyze penguin species and identify patterns in physical characteristics such as body mass and flipper length.


```python
import pandas as pd
import seaborn as sns

# Load dataset
df = sns.load_dataset("penguins")

df.head()
```

```python
import matplotlib.pyplot as plt

# Average body mass by species
body_mass = df.groupby('species')['body_mass_g'].mean()

plt.figure()
body_mass.plot(kind='bar')
plt.title("Average Body Mass by Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

### Insight: Different penguin species have noticeably different average body masses, suggesting biological and environmental differences between species.
```

```python
import matplotlib.pyplot as plt

# Average flipper length by species
flipper_length = df.groupby('species')['flipper_length_mm'].mean()

plt.figure()
flipper_length.plot(kind='bar')
plt.title("Average Flipper Length by Species")
plt.xlabel("Species")
plt.ylabel("Flipper Length (mm)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

### Insight: Penguin species show clear differences in flipper length, which may relate to variations in swimming ability and environmental adaptation.
```

```python
import matplotlib.pyplot as plt

# Histogram of body mass
plt.figure()

plt.hist(df['body_mass_g'].dropna(), bins=20)

plt.title("Distribution of Penguin Body Mass")
plt.xlabel("Body Mass (g)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

### Insight: The distribution of body mass shows a range of sizes across penguins, with most individuals clustering within a central range, indicating natural variation within species.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure()
sns.boxplot(x='species', y='body_mass_g', data=df)

plt.title("Body Mass Distribution by Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")

plt.tight_layout()
plt.show()

### Insight: The boxplot reveals the spread and variability of body mass within each species, showing not only average differences but also the presence of outliers and distribution ranges.
```

```python
import seaborn as sns

# Select numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure()
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Between Penguin Features")

plt.tight_layout()
plt.show()

### Insight: The correlation heatmap shows relationships between numerical features, such as body mass and flipper length, indicating that larger penguins tend to have longer flippers.
```
